```python
import random
import numpy as np
import matplotlib.pyplot as plt

def extinction(P,m,T = 10):
    """ Function checks whether population will be dead or alive in specific time steps till T time step.
    -------------------------------------------------
    P - list of probabilites of having i children.
    m - number of men in first generation
    T - maximum time step
    -------------------------------------------------
    1 - dead
    0 - alive
    """
    deadOrAlive = []
    for _ in range(T):
        newChildren = 0
        for _ in range(m):
            numberOfSons = random.choices(range(len(P)), weights = P)
            newChildren += numberOfSons[0]
        m = newChildren
        if m <= 0:
            deadOrAlive.append(1)
        else:
            deadOrAlive.append(0)
    return deadOrAlive

def probabilityOfExtinction(P, m, T = 10, n = 100):
    """ Functions checks probability of extinction after T time steps.
    -------------------------------------------------
    P - list of probabilites of having i children.
    m - number of men in first generation
    T - maximum time step
    n - number of simulations
    -------------------------------------------------
    """
    probabilites = np.array([0]*T)
    for _ in range(n):
        deadOrAlive = extinction(P,m,T)
        probabilites += np.array(deadOrAlive)
    return probabilites/n
    
def plotGeneration(family):
    """ Function plots distribution function. """
    plt.ylim([0,1])
    plt.ylabel('Probability of extinction')
    plt.xlabel('Generation')
    plt.plot(family)
    plt.show() 
```

```python
Pf = [0.4, 0.1, 0.2, 0.3]
Mf = 3
extinction(Pf,Mf)
```

```python
Fabiusze = probabilityOfExtinction(Pf,Mf,20)
print(Fabiusze)
```

```python
plotGeneration(Fabiusze)
```

```python
Pe = [0.3, 0.5, 0.05, 0.05, 0.05, 0.05]
Me = 5
Emiliusze = probabilityOfExtinction(Pe, Me, 30)
plotGeneration(Emiliusze)
```

```python
Pk = [0.5, 0.5]
Mk = 20
Korneliusze = probabilityOfExtinction(Pk,Mk,20)
plotGeneration(Korneliusze)
```
