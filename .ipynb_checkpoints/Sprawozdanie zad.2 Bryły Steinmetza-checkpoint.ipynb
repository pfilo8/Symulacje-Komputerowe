```python
import random
import matplotlib.pyplot as plt
import numpy as np
import time
from scipy.optimize import fmin, fminbound

plt.style.use('ggplot')
```

```python
def generation(x_vector, a_vector, p_vector, m, value=1):
    if value == 1:
        x_n = x_vector[0]
        for i in range(1, len(x_vector) - 1):
            x_n += a_vector[i]*x_vector[i]**p_vector[i]
        return int(x_n % m)
    result = []
    for _ in range(value):
        x_n = x_vector[0]
        for i in range(1, len(x_vector) - 1):
            x_n += a_vector[i]*x_vector[i]**p_vector[i]
        x_n = int(x_n % m)
        result.append(x_n)
        for i in range(len(x_vector)-1):
            x_vector[i] = x_vector[i+1]
        x_vector[-1] = x_n
    
    return result
```

```python
def uniform_generator(a=0, b=1, value=1):
    out = []
    m = 1/(5**(-303))
    a_vector = [m*2**87, m*3**64, m*5**76]
    p_vector = [0.05738, 1.5646, 1.23456]
    for _ in range(value):
        x_n = [time.clock()%400, time.clock() % 400, time.time()]
        if value == 1:
            return a + generation(x_n, a_vector, p_vector, m)/(m-1)*(b-a)
        out.append(a + generation(x_n, a_vector, p_vector, m)/(m-1)*(b-a))
    return out
```

```python
def policz(d,t = 10000):
    count = 0
    for _ in range(t):
        wylosowaneLiczby = [random.random() for _ in range(d)]
        y = random.uniform(0,d)
        if y <= sum(wylosowaneLiczby):
            count += 1
    return d*count/t
```

```python
def integral(f, start=0, end=1, repeat=10**5):
    f_min = min(0, f(fminbound(f, start, end)))
    f_max = max(0, f(fminbound(lambda x: -f(x), start, end)))
    s = 0
    for i in range(repeat):
        x = uniform_generator(start, end)
        y = uniform_generator(f_min, f_max)
        if 0 <= y <= f(x):
            s += 1
        elif f(x) <= y <= 0 :
            s -= 1
    return s/repeat * (end - start) * (f_max - f_min)
```

```python
def steinmetz(r, fun = lambda x: 16 * x**2):
    v = integral(fun, end = r, repeat = 10**4)
    return v
```

```python
def proportion(r, f):
    return steinmetz(r, f)/f(r)
```

# 1. Wstęp teoretyczny

W tym zadaniu będziemy chcieli symulacyjnie wyznaczyć stosunek pola powierzchni do objętości 2- i 3- Bryły Steinmetza. W tym celu wykorzystamy uogólnioną metodę Monte Carlo do szacowania całek funkcji d-zmiennych postaci:
$$\idotsint_D f(x_1,\dots,x_n) \,dx_1 \dots dx_n,$$
gdzie $f:\mathbb{R}^n  \rightarrow  \mathbb{R}$, a $ D \subset \mathbb{R}^n$ jest zbiorem ograniczonym.

Do przetestowania poprawności metody oraz szybkości zbieżności algorytmu wykorzystamy całkę:

$$\idotsint_D x_1+x_2+\dots+x_n dx_1dx_2\dots dx_n, $$

gdzie $D = [0,1]^n$

Obliczmy tą całkę analitycznie, aby móc sprawdzić jaki błąd otrzymujemy podczas symulacji:

$$\idotsint_D x_1+x_2+\dots+x_n dx_1dx_2\dots dx_n = \idotsint_D x_1 dx_1dx_2\dots dx_n + \dots + \idotsint_D x_n dx_1dx_2\dots dx_n \stackrel{(*)}= n* \frac{1}{2} = \frac{n}{2}, $$

gdzie 
$$(*) = \idotsint_0^1 x_n dx_1dx_2\dots dx_n = \int_0^1x_ndx_n = \frac{x^2}{2}\Big|_0^1 = \frac{1}{2} $$

# 2. Rozwiązanie

```python
ts = list(map(int,(np.logspace(1,4,10))))

plt.figure(figsize = (10,6))
maxd = 6

for d in range(1,maxd):
    score = []
    for t in ts:
        temp = []
        for _ in range(30):
            temp.append(abs(policz(d,t) - d*0.5))
        score.append(np.mean(temp))
    plt.plot(ts,score,label = str(d), color = (0,1 - d/maxd,1 - d/maxd))
    plt.scatter(ts,score, color = (0,1 - d/maxd,1 - d/maxd))

plt.title('Porównanie szybkości zbieżności całek')
plt.xscale('log')
plt.legend(loc = 'best')
plt.show()
```

Powyższy wykres przedstawia porównanie szybkości zbieżności całek w zależności od wymiaru (kolor punktów oraz linii) całki oraz długości symulacji próby. Możemy zauważyć, że wraz ze wzrostem wymiaru przestrzeni, po której całkujemy, wzrasta czas oczekiwania na zbieżność całki. Ponadto widzimy, że dla długości symulacji próby równej ok. $10^4$ kroków Monte Carlo osiągamy już bardzo zadowalającą zbieżność.

# 3. Objętość bryły Steinmetz'a

Celem tego zadania jest sprawdzenie stosunku pola powierzchni kuli, 2-brył Steinmentza oraz 3-brył Steinmetza do odpowiednich objętości. Zgodnie z treścią zadania możemy wykorzystać analityczne wzory na pola powierzchnii, natomiast objętości należy wysymulować metodą Monte Carlo. Przedstawmy zatem wzory opisujące badane zależności.

**Kula**

Pole powierzchni =

Objętość = Całkowa = Analityczna

**2-bryła Steinmetza**

Pole powierzchni =

Objętość = Całkowa = Analityczna

**3-bryła Steinmetza**

Pole powierzchni =

Objętość = Całkowa = Analityczna

Można łatwo zauważyć, że dla wszystkich przypadków stosunek $\frac{Pole powierzchni}{Objętość} = \frac{3}{r}$

Na podstawie:
- https://en.wikipedia.org/wiki/Steinmetz_solid
- http://mathworld.wolfram.com/SteinmetzSolid.html

```python
ObjetoscSymulacyjna = []
ObjetoscAnalityczna = []
stosunek = []
zakres = list(range(1,11))
zakresAnalityczny = np.linspace(min(zakres), max(zakres), 100)
stosunekAnalityczny = [3/r for r in zakresAnalityczny]

for r in zakres:
    S = 4*np.pi*r**2
    Vs = 2*np.pi * integral(start = 0, end = r, f = lambda x: r**2 - x**2, repeat = 10**5)
    Va = 4*np.pi*r**3/3
    ObjetoscSymulacyjna.append(Vs)
    ObjetoscAnalityczna.append(Va)
    stosunek.append(S/Vs)

plt.figure(figsize = (14,8))
    
plt.subplot(1,2,1)
minimum = min(min(ObjetoscAnalityczna),min(ObjetoscSymulacyjna))
maximum = max(max(ObjetoscAnalityczna),max(ObjetoscSymulacyjna))
plt.xlabel('Objetosc bryly analityczna')
plt.ylabel('Objetosc bryly symulacyjna')
plt.title('Porownanie objetosci analitycznej z symulacyjna ')
plt.scatter(ObjetoscAnalityczna, ObjetoscSymulacyjna)
plt.plot([minimum, maximum], [minimum, maximum], 'k--', alpha = 0.5)

plt.subplot(1,2,2)
plt.xlabel('r')
plt.ylabel('Stosunek pola powierzchnii do objętości')
plt.title('Kula')
plt.scatter(zakres, stosunek)
plt.plot(zakresAnalityczny, stosunekAnalityczny, 'k--', alpha = 0.5)
plt.show()
```

```python
ObjetoscSymulacyjna = []
ObjetoscAnalityczna = []
stosunek = []
zakres = list(range(1,11))
zakresAnalityczny = np.linspace(min(zakres), max(zakres), 100)
stosunekAnalityczny = [3/r for r in zakresAnalityczny]

for r in zakres:
    S = 16*r**2
    Vs = integral(start = -r, end = r, f = lambda x: (2*np.sqrt(r**2 - x**2))**2, repeat = 10**5)
    Va = 16*r**3/3
    ObjetoscSymulacyjna.append(Vs)
    ObjetoscAnalityczna.append(Va)
    stosunek.append(S/Vs)

plt.figure(figsize = (14,8))
    
plt.subplot(1,2,1)
minimum = min(min(ObjetoscAnalityczna),min(ObjetoscSymulacyjna))
maximum = max(max(ObjetoscAnalityczna),max(ObjetoscSymulacyjna))
plt.xlabel('Objetosc bryly analityczna')
plt.ylabel('Objetosc bryly symulacyjna')
plt.title('Porownanie objetosci analitycznej z symulacyjna ')
plt.scatter(ObjetoscAnalityczna, ObjetoscSymulacyjna)
plt.plot([minimum, maximum], [minimum, maximum], 'k--', alpha = 0.5)

plt.subplot(1,2,2)
plt.xlabel('r')
plt.ylabel('Stosunek pola powierzchnii do objętości')
plt.title('2-bryła Steinmetza')
plt.scatter(zakres, stosunek)
plt.plot(zakresAnalityczny, stosunekAnalityczny, 'k--', alpha = 0.5)
plt.show()
```

```python
ObjetoscSymulacyjna = []
ObjetoscAnalityczna = []
stosunek = []
zakres = list(range(1,11))
zakresAnalityczny = np.linspace(min(zakres), max(zakres), 100)
stosunekAnalityczny = [3/r for r in zakresAnalityczny]

for r in zakres:
    S = 3*(16 - 8*np.sqrt(2))*r**2
    Vs = 6 * integral(start = r/np.sqrt(2), end = r, f = lambda x: (2*np.sqrt(r**2 - x**2))**2, repeat = 10**5) + (np.sqrt(2)*r)**3
    Va = (16 - 8*np.sqrt(2))*r**3
    ObjetoscSymulacyjna.append(Vs)
    ObjetoscAnalityczna.append(Va)
    stosunek.append(S/Vs)

plt.figure(figsize = (14,8))
    
plt.subplot(1,2,1)
minimum = min(min(ObjetoscAnalityczna),min(ObjetoscSymulacyjna))
maximum = max(max(ObjetoscAnalityczna),max(ObjetoscSymulacyjna))
plt.xlabel('Objetosc bryly analityczna')
plt.ylabel('Objetosc bryly symulacyjna')
plt.title('Porownanie objetosci analitycznej z symulacyjna ')
plt.scatter(ObjetoscAnalityczna, ObjetoscSymulacyjna)
plt.plot([minimum, maximum], [minimum, maximum], 'k--', alpha = 0.5)

plt.subplot(1,2,2)
plt.xlabel('r')
plt.ylabel('Stosunek pola powierzchnii do objętości')
plt.title('3-bryła Steinmetza')
plt.scatter(zakres, stosunek)
plt.plot(zakresAnalityczny, stosunekAnalityczny, 'k--', alpha = 0.5)
plt.show()
```

Z powyższej analizy widać, że wzory na objętość przedstawione w części teoretycznej tego zadania pokrywają się ze obliczonymi całkami. Ponadto widać, że wszystkie 3 bryły idealnie spełniają zależność pola powierzchnii do objętości jako $\frac{3}{r}$. Jest to całkiem interesujący fakt, że wszystkie te 3 brył spełniają taką zależność i można by było przeprowadzić dalsze badania, czy to tylko przypadek dla tych brył, czy istnieje też taka zależność dla większej klasy brył.
