```python
import random
import math
import matplotlib.pyplot as plt
```

# I metoda generowania processu Poissona

```python
def procesPoissona(lam,T):
    """Funkcja generująca czasy oczekiwania na skok w procesie Poissona
    lam - intensywnosc procesu
    T - czas trwania procesu
    ---------------------------
    S - momenty skoków
    """
    I = 0
    t = 0
    S = [0]
    while t <= T:
        u = random.random()
        t -= math.log(u)/lam
        I += 1
        S.append(t)
    return S
```

```python
def plotProcesPoissona(S, color = 'k'):
    N = range(len(S))
    plt.scatter(S[1:-1], N[:-2], s = 10, facecolors='none', edgecolors='k')
    # Wycinki wynikają z przesunięcia punktów oraz braku otwartego kółka na końcu.
    for i in range(len(S) - 1):
        x = [S[i], S[i+1]]
        y = [N[i], N[i]]
        plt.plot(x,y, color = color)
    plt.plot(S[-1], N[-1])
```

```python
plt.figure(figsize = (12,10))
plt.xlabel('t')
plt.ylabel('N(t)')
for i in range(5):
    S = procesPoissona(1,100)
    plotProcesPoissona(S, (0, i/5, 0))
plt.show()
```

# II metoda generowania procesu Poissona

```python
import numpy as np
```

```python
def procesPoissona2(lam, T):
    n = np.random.poisson(lam*T)
    if n == 0:
        return None
    U = np.random.uniform(0,T,n)
    U.sort()
    return U
```

```python
U = procesPoissona2(1, 100)
```

```python
U
```

```python
plt.figure(figsize = (12,10))
plt.xlabel('t')
plt.ylabel('N(t)')
for i in range(5):
    S = procesPoissona2(1,100)
    plotProcesPoissona(S, (0, 0, i/5))
plt.show()
```

```python
import time
```

```python
def countTime(f, lam, T, t):
    start = time.clock()
    for _ in range(t):
        f(lam, T)
    return time.clock() - start
```

Czas wykonania 100000 procesów Poissona pierwszą metodą:

```python
countTime(procesPoissona, 1, 20, 100000)
```

Czas wykonania 100000 procesów Poissona drugą metodą:

```python
countTime(procesPoissona2, 1, 20, 100000)
```

Druga metoda jest zdecydowanie szybsza.
Wadą pierwszej metody jest użycie pętli while, która uniemożliwa wykorzystanie szybkich operacji macierzowych z biblioteki numpy, co ma miejsce w drugiej metodzie. Natomiast wadą drugiej metody jest konieczność posortowania otrzymanej tablicy, co jest czasochłonne.

Pierwsz metoda nadaje się do rzeczywistej symulacji procesu Poissona.

```python
def animowanyProcesPoissona(lam,T):
    """
    lam - intensywnosc procesu
    T - czas trwania procesu
    ---------------------------
    S - momenty skoków
    """
    plt.figure(figsize = (12,10))
    plt.xlabel('t')
    plt.ylabel('N(t)')
    I = 0
    t = 0
    S = [0]
    while t <= T:
        u = random.random()
        t -= math.log(u)/lam
        I += 1
        S.append(t)
    return S
```
