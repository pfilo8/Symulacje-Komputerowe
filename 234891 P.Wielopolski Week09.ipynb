# Zadanie 32 - II metoda

```python
def podwojnyPoisson2(lam, T):
    lamMax = max([lam(t) for t in range(1, T)])
    t = 0.01
    S = [0]
    while t <= T:
        u1 = random.random()
        t -= math.log(u1)
        u2 = random.random()
        if u2 <= lam(t)/lamMax:
            S.append(t)
    return S
```

```python
def plotProcesPoissona(S, color = 'k'):
    N = list(range(len(S)))
    S.append(S[-1] + 1)
    N.append(N[-1])
    plt.scatter(S, N, s = 10, facecolors='none', edgecolors='k')
    # Wycinki wynikają z przesunięcia punktów oraz braku otwartego kółka na końcu.
    for i in range(len(S) - 1):
        x = [S[i], S[i+1]]
        y = [N[i], N[i]]
        plt.plot(x,y, color = color)
    plt.plot(S[-1], N[-1])
```

```python
def podwojnyPoisson2wrapper(f, T):
    try:
        wynik1 = podwojnyPoisson2(f, T)

        plotProcesPoissona(wynik1)
        plt.ylabel('$N(t)$')
        plt.xlabel('$t$')
        plt.show()
    except Exception as e:
        print(e)
```

```python
def f1(t):
    return max(0, math.log(random.gauss(2, 1/t)))
```

```python
def f2(t):
    return math.exp(random.gauss(0, t**2))
```

```python
def f3(t):
    return random.expovariate((math.sin(t))**2)*math.exp(-np.random.standard_cauchy())
```

```python
def f4(t):
    return abs(random.gauss(0,1))
```

# Zadanie 33 - Geometryczny Ruch Browna

```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generujGeometrycznyRuchBrowna(mu, sigma, S0, T, r):
    ilosc_krokow = math.ceil(T/r)
    B = 0
    S = [0]
    for t in range(ilosc_krokow):
        B += math.sqrt(r)*random.gauss(0,1)
        S.append(S0*math.exp((mu - (sigma**2)/2)*t + sigma*B))
    return S
```

```python
def ruchBrowna(T):
    return np.cumsum(np.random.randn(T))
```

```python
def geometrycznyRuchBrowna(mu, sigma, S0, T):
    B = ruchBrowna(T)
    t = np.arange(0,T,1)
    return S0*np.exp((mu - (sigma**2)/2)*t + sigma*B)
```

```python
S = geometrycznyRuchBrowna(0.55, 1, 50,200)
plt.plot(S)
plt.show()
```

```python
S = generujGeometrycznyRuchBrowna(0.55, 1, 5, 10, 0.05)
```

# Zadanie 32 - I metoda

```python
def procesPoissona(lam,T):
    """Funkcja generująca czasy oczekiwania na skok w procesie Poissona
    lam - intensywnosc procesu
    T - czas trwania procesu
    ---------------------------
    S - momenty skoków
    """
    I = 0
    t = 0
    S = [0]
    while t <= T:
        u = random.random()
        t -= math.log(u)/lam
        I += 1
        S.append(t)
    return S
```

```python
def procesCoxa(fLam, T):
    temp = [0] + [fLam(t) for t in range(1,T)]
    lambdaDuza = np.cumsum(np.array(temp))
    JednorodnyPoisson = procesPoissona(1, max(lambdaDuza))

    procesCoxa = []

    for el in lambdaDuza:
        for i in range(len(JednorodnyPoisson)):
            if JednorodnyPoisson[i] > el:
                procesCoxa.append(i-1)
                break
    return lambdaDuza.tolist(),procesCoxa
```

```python
def plotProcesCoxa(fLam, T):
    procesLambda, procesCoxa1 = procesCoxa(fLam, T)
    plt.figure(figsize = (10,6))

    plt.subplot(1,2,1)
    plotProcesPoissona(procesLambda)

    plt.subplot(1,2,2)
    plotProcesPoissona(procesCoxa1)

    plt.show() 
```
