```python
def BoxCox(data, lam):
    """
    data - numpy.array - szereg czasowy
    lam - współczynnik trasformacji Boxa-Coxa
    ------------------------------------------
    Dane po trasformacji Boxa-Coxa
    """
    if lam == 0:
        return np.log(data)
    else:
        return (data**(lam) - 1)/lam
```

```python
def Differencing(data, h = 1):
    """
    data - numpy.array - szereg czasowy
    h - opóźnienie
    ----------------------------------
    Dane po jednokrotnej operacji różnicowanie z opóźnieniem h.
    """
    return data[h:] - data[:-h]
```

```python
def ExponentialSmoothing(data, alpha):
    """
    data - numpy.array - szereg czasowy
    alpha - współczynnik wygładzania
    ----------------------------------
    Dane po prostym wygładzaniu wykładniczym z parametrem alpha.
    """
    S = [data[0]]
    for i in range(1, len(data)):
        S.append((1-alpha)*S[-1] + alpha*data[i])
    return np.array(S)
```

```python
def ARMAsimulation(n, p_coefs, q_coefs):
    
    """
    Zwraca wektor n-q+p+1 wysymulowanych wartości modelu autoregresji ruchomej średniej.
    
    Parameters:
    -----------
    n : int
        długość wektora
    p_coefs : array-like
        współczynniki modelu AR w kolejności t1, ..., tp
    q_coefs : array-like
        współczynniki modelu MA w kolejności z1, ..., zq
    """
    phi = np.flip(p_coefs, axis = 0)
    p = len(phi)
    
    theta = [1] + q_coefs
    theta = np.flip(theta, axis = 0)
    q = len(theta)

    X = np.random.randn(p)
    Z = np.random.randn(n)

    for i in range(n-q):
        X = np.append(X, np.sum(X[i:i+p]*phi) + np.sum(Z[i:i+q]*theta))

    return X
    
```

```python
def MAsimulation(n, coefs):
    """
    Zwraca wektor n-q wysymulowanych wartości ruchomej średniej.
    
    Parameters:
    -----------
    n : int
        długość wektora
    coefs : array-like
        współczynniki modelu MA w kolejności t1, ..., tq
    """
    return ARMAsimulation(n, [], coefs)
```

```python
def ARsimulation(n, coefs):
    """
    Zwraca wektor n wysymulowanych wartości modelu autoregresji.
    
    Parameters:
    -----------
    n : int
        długość wektora
    coefs : array-like
        współczynniki modelu AR w kolejności t1, ..., tq
    """
    return ARMAsimulation(n, coefs, [])
```

# Szeregi stacjonarne

Bardzo ważną rodzinę szeregów czasowych stanowią tak zwane szeregi stacjonarne. W szczególności, gdy w dalszej części rozdziału będziemy zajmować się modelami ARMA, szczególnie ważne w dopasowywaniu danych będzie powyższa cecha. Podajmy zatem własności, które musi spełniać szereg, aby można było go nazwać stacjonarnym.

- $Corr(X_r,X_s) = Corr(X_{r+t},X_{s+t}) = \rho(s-r) = \rho(h)$ - korelacja między obserwacjami zależy tylko od odstępu czasowego h (opóźnienie)
- $Var(X_t) = const$ - niezmienna w czasie wariancja
- $EX_t = const $ - stała wartość oczekiwana (brak składowych deterministycznych, tzn. trendów długoterminowych lub sezonowych)

# Autokowariancja oraz autokorelacja

Autokowariancja, to wielkość opisująca kowariancję danego procesu z tym samym procesem przesuniętem o pewien odcinek czasu $\tau$. Matematycznie możemy zapisać to w następujący sposób:

$$ K_{xx}(\tau) = E(X_t - \mu_{t})(X_{t+\tau} - \mu_{t+\tau}) = E(X_tX_{t+\tau}) - \mu_t\mu_{t+\tau},$$
gdzie $\mu_t = EX_t$
Często zamiast autokowariancji stosujemy autokorelację, czyli unormowaną funkcję autokowariancji, która opisuje korelację Pearsona pomiędzy procesem w chwili $t$ a procesem w chwili $s$. Definiujemy ją w następujący sposób:

$$ R(\tau) = \frac{K_{xx}(\tau)}{\sigma^2} = \frac{E(X_t - \mu_{t})(X_{t+\tau} - \mu_{t+\tau})}{\sigma^2}, $$
gdzie $\sigma^2 = D^2X_t$

Dla procesów ze znaną wariancją oraz odchyleniem standardowym możemy skorzystać z następującego estymatora dla funkcji autokorelacji:
$$ \hat{R}(\tau)=\frac{1}{(n-\tau) \sigma^2} \sum_{t=1}^{n-\tau} (X_t-\mu)(X_{t+\tau}-\mu) $$
W tym przypadku otrzymujemy estymator nieobciążony. W wypadku, gdy nie znamy średniej i wariancji, możemy zastąpić je przez wartości prókowe. Niestety w tym przypadku nasz estymator będzie obciążony.

## ACF (AutoCorrelation Function)

**Def. Funkcja autokorelacji ACF(h)**

Funkcję autokorelacji ACF definiujemy jako
$$ ACF(h) = \frac{\gamma(h)}{\gamma(0)}, h = 0,1,2,\dots,n-1, $$
gdzie $\gamma(h)$ jest próbkową funkcją autokowariancji
$$ \gamma(h) = \frac{1}{n} \sum_{j=1}^{n-h}(X_{j+h}-\overline{X})(X_{j}-\overline{X}), h=0,1,2,\dots,n-1, $$
gdzie $\overline{X} = (X_1 + X_2 + \dots + X_n)/n$.

    
Funkcja autokorelacji jest miarą zależności (korelacji liniowej) między obserwacjami oddalonymi o $h$ jednostek czasowych. Parametr h jest tzw. parametrem opóźnienia (ang. *lag*). ACF jest uogólnieniem korelacji liniowej, tzn. funkcja ACF jest oszacowaniem autokorelacji $\rho(h) = Corr(X_{t+h},X_t)$.
    
Reguły stosowane przy interpretacji ACF:
- Dodatnie i powoli zanikające wartości funkcji ACF  sugerują trend w danych
- Bardzo powoli zanikające oraz cykliczne warości ACF sugerują obecność sezonowości
- Szybkość zanikania ACF pozwala nam na identyfikowanie klas modeli SCz (do dopisania jeszcze)

```python
plot_acf(data.Zamkniecie, lags = np.arange(50), unbiased = True, alpha = 0.05)
plt.show()
```

Przykładowy wykres autokorelacji dla naszych danych z giełdy. Stosując powyżej opisane reguły możemy stwierdzić, że nasze dane zawierają bardzo duży trend.

## PACF (Partial AutoCorrelation Function)

**Def. Funkcja autokorelacji cząstokowej (PACF)**
Funkcja próbkowej autokorelacji cząstkowej (PACF) jest oszacowaniem teoretycznyej funkcji autokorelacji cząstkowej, określanej jako
$$ \alpha(1) = Corr(X_2,X_1) = \rho(1) $$
oraz
$$ \alpha(h) = Corr(X_{h+1} - P_{\overline{sp}\{1,X_2,\dots,X_h\}}X_{h+1},X_1 - P_{\overline{sp}\{1,X_2,\dots,X_h\}}X_1) dla h\ge2, $$
gdzie $P_{\overline{sp}\{1,X_2,\dots,X_h\}}$ oznacza rzut ortogonalny na podprzestrzeń liniową wyznaczoną przez $1,X_2,\dots,X_h$.

PACF mierzy zależność pomiędzy obserwacjami odległymi o h jednostek czasu, tzn. $X_t, X_{t+h}$. W przeciwieństwie do ACF, PACF eliminuje zależności między $X_t$ oraz $X_{t+h}$, które pochodzą z obserwacji pośrednich $X_{t+1},\dots,X_{t+h-1}$.

Reguły stosowane przy interpretacji PACF:
- Duża wartość (bliska 1) dla opóźnienia równego 1 może świadczyć o silnym trendzie wzrostowym w danych
- Jest to użyteczne narzędzie do identyfikacji modeli autoregresyjnych

```python
plot_pacf(data.Zamkniecie, lags = 50, alpha = 0.05)
plt.show()
```

Ponownie korzystając z reguł stosowanych przy interpretacji PACF możemy zauważyć, że nasze zawierają silny trend wzrostowy, co również można zobaczyć na pierwszym wykresie zależności ceny od czasu.

# Modele stacjonarne AR,MA,ARMA

W tej części sprawozdania zajmiemy się opisanie podstawowych modeli stacjonarnych. Są one jednymi z najlepiej zbadanych teoretycznie modeli dla szeregów czasowych. Są one często stosowane przez praktyków oraz stanowią podstawę do budowania bardziej zaawansowanych modeli.

### Biały szum ($WN(0,\sigma^2$))

Białym szumem o średniej 0 i wariancji $\sigma^2$ (oznaczanym $Z_t \sim WN(0,\sigma^2$)) nazywamy ciąg zmiennych nieskorelowanych o jednakowym rozkładzie, dla którego wartość średnia równa się 0, a wariancja równa się $\sigma^2$, tzn.
- $Corr(Z_t,Z_{t+h})=0 $ dla dowolnego $t$ i $h\ne0$
- $EZ_t = 0$ i $VarZ_t = \sigma^2$ dla dowolnego $t$

```python
plt.figure(figsize = (8,2))

plt.plot(np.random.randn(1000))
plt.title('Symulacja białego szumu $WN(0,1)$')
plt.show()
```

### Model ruchomej średniej ($MA(q)$)

Modelem ruchomej średniej rzędu q nazywamy stacjonary szereg czasowy $X_t$ spełaniający równanie
$$ X_t = Z_t + \theta_1Z_{t-1} + \dots + \theta_qZ_{t-q}, $$
gdzie $Z_t$ - szereg typu biały szum

```python
MAwywolanie()
```

### Model autoregresji ($AR(p)$)

Modelem autoregresji rzędu p nazywamy stacjonarny szereg czasowy $X_t$ spełniający równanie
$$ X_t = \phi_1X_{t-1} + \dots + \phi_pX_{t-p} + Z_t, $$
gdzie $Z_t$ - szereg typu biały szum

```python
ARwywolanie()
```

### Model autoregresji ruchomej średniej ($ARMA(p,q)$)
Modelem autoregresji ruchomej średniej nazywamy stacjonarny szereg czasowy $X_t$ spełniający równanie
$$ X_t = \phi_1X_{t-1} + \dots + \phi_pX_{t-p} + Z_t + \theta_1Z_{t-1} + \dots + \theta_qZ_{t-q}, $$
gdzie $Z_t$ - szereg typu biały szum

# Modele niestacjonarne ARIMA, SARIMA

Przed wprowadzeniem modeli niestacjonarnych będziemy potrzebowali dwóch dodatkowych transformacji:

### Różnicowanie

Różnicowanie jest to jedna z podstawowych i najczęściej wykorzystywanych transformacji. Stosujemy ją między innymi w celu przekształcenia danych do postaci stacjonarnej. Dzięki niej możemy usunąć z danych trend liniowy, kwadratowy lub wielomianowy, a także trend sezonowy.
Podstawową wersją różnicowania jest różnicowanie z opóźnieniem 1. W tym przypadku z wektora danych $X_1, X_2, \dots, X_n$ wyznaczamy wektor $X_2-X_1, X_3-X_2, \dots, X_n-X_{n-1}$. Otrzymujemy w ten sposób nowy szereg postaci $Y_t = X_t - X_{t-1}$, gdzie $t = 2,3,\dots,n$. Zauważmy, że nowy szereg zawiera $n-1$ wartości, ponieważ nie jest możliwe wyznaczenie różnicy $Y_1$.

```python
roznicowanieWywolanie()
```

Jak widzimy, po zróżnicowaniu naszego szeregu wyeliminowaliśmy trend długoterminowy. W zróżnicowanych danych możemy jednak zauważyć wzrost wariancji z czasem. Formalnie można udowodnić, że różnicowanie z opóźnieniem 1 powoduje między innymi wyeliminowanie z szeregu deterministycznego trendu liniowego.

Operację różnicowania możemy zdefiniować w bardziej formalny sposób:

**Def.** Operator różnicowania z opóźnieniem 1 definiujemy jako
$$ \nabla X_t = X_t - X_{t-1} = (1-B)X_t,$$
gdzie $B$ jest tzw. operatorem przesunięcia wstecz (*ang. backward shift operator*) zdefuniowany jako $BX_t = X_{t-1}$.

### Różnicowanie z opóźnieniem sezonowym

Różnicowanie z opóźnieniem 1 pomaga nam wyeliminować trend długoterminowy, lecz niestety jest nieskuteczne w przypadku występowania sezonowości. Aby było to możliwe, musimy wykorzystać różnicowanie z opóźnieniem sezownowym $s$, które w praktyce oznacza wyznaczenie szeregu $Y_t = X_t - X_{t-s}$, gdzie $t = s+1, \dots, n$. Warto zauważyć, że nasz szereg posiada wtedy $n-s$ obserwacji.

Aby formalnie zdefiniować operację różnicowania z opóźnieniem sezonowym, zauważmy prostą własność operatora $B$:

$$ B^S(X_t) = B(B^{s-1}X_t) = X_{t-s} $$

Wprowadźmy zatem formalną definicję:

**Def.** Dla ustalonego opóźnienia $s$ operator różnicowania definiujemy jako:

$$ \nabla_sX_t = X_t - X_{t-s} = (1-B^s)X_t $$

Praktycznie, żeby usunąć z danych miesięcznych trend sezonowy należy wykorzystać różnicowanie z opóźnieniem sezonowym 12, natomiast dla danych kwartalnych z opóźnieniem sezonowym 4. Co ważne, w wyniku tego typu różnicowania usuwamy także trend liniowy z danych.

Mając tak zdefiniowane dwie transformacje możemy wprowadzić kolejne dwa modele:

### Model $ARIMA(p,d,q)$

Szereg $X_t$ nazywamy procesem $ARIMA(p,d,q)$ , jeżeli po d-krotnym ($d\ge0$) zróżnicowaniu z opóźnieniem 1 jest już procesem $ARMA(p,q)$, tzn. $X_{t}$ spełnia równianie
$$ \phi^{*}(B)X_t = \phi(B)(1-B)^dX_t = \theta(B)Z_{t},$$

gdzie $Z_t \sim WN(0,\sigma^2)$ oraz $\phi(z)$,$\theta(z)$ są wielomianami odpowiednio stopnia $p$ i $q$ oraz $\phi(z)\ne0$ dla $|z|\le1$


Zauważmy, że $ARIMA(p,0,q) = ARMA(p,q)$. Zatem modele ARIMA stanowią ogólnienie klasy ARMA obejmujące szeregów niestacjonarnych.

### Model $SARIMA(p,d,q)(P,D,Q)_s)$

Dla $d,D\ge0$ i $d,D\in N$, proces $X_t$ jest procesem $SARIMA(p,d,q)(P,D,Q)_s$ o okresie $s$ (inaczej sezonowym modelem $ARIMA(p,d,q)(P,D,Q)_s$ o okresie $s$), jeżeli zróżniocwany szereg $Y_t = (1-B)^d(1-B^s)^DX_t$ jest modelem $ARMA(p,q)(P,Q)_s$, zdefiniowanym w następujący sposób

$$ \theta(B)\Phi(B^s)Y_t = \theta(B)\Theta(B^s)Z_t,$$

gdzie $Z_t \sim WN(0,\sigma^2)$, $\phi(z) = 1 - \phi_1z - \dots - \phi_pz^p, \Phi(z) = 1 - \Phi_1z - \dots - \Phi_Pz^P, \theta(z) = 1+ \theta_1z + \dots + \theta_qz^q, \Theta(z) = 1+ \Theta_1z + \dots + \Theta_Qz^Q.$


Zauważmy, że $SARIMA(p,d,q)(P,D,Q)_s)$ zawiera wszystkie wcześniej poznane modele. Dla przykładu $ARIMA(p,d,q) = SARIMA(p,d,q)(0,0,0)_s)$.

# Transformacje

Niestety w zastosowaniach najczęściej spotykamy się z niestacjonarymi szeregami czasowymi. W tym celu często korzystamy z odpowiednich transformacji, które sprowadzają nasz model do postaci stacjonarnej. Do tej pory opisaliśmy różnicowanie oraz różnicowanie sezonowe. Oto przykłady dwóch innych przydatnych transformacji.

## Transformacja Boxa-Coxa

Kiedy korzystamy z trasformacji Boxa-Coxa:
- występowanie wzrostu (lub spadku) wahań amplitudy w kolejnych okresach
- niejednorodna zmienność danych w kolejnych okresach
- dane odstające mające duży wpływ na szereg
- chęć uzyskania rozkładów bardziej zbliżonych do normalnych (metody analityczne często wymagają normalności)



Trasformacja Boxa-Coxa to rodzina transformacji zdefiniowana jako
$$
f_{\lambda}(x) = \left\{ \begin{array}{ll}
\frac{x^{\lambda}-1}{\lambda} & \textrm{dla $\lambda\ne0$ }\\
ln(x) & \textrm{dla $\lambda = 0$}\\
\end{array} \right.
$$

Do najczęściej stosowanych transoformacji wartości  należą 
- $\lambda = \frac{1}{2}$ - przekształcenie pierwiastkowe
- $\lambda = 0$ - przekształcenie logarytmiczne

Często stosowane praktyką jest dobór odpowiedniej wartości $\lambda$ na podstawie wykresu danych przed i po transformacji. W szczególności wybieramy taką wartość $\lambda$, który przynosi najlepszą stabilizację wariancji.

Metoda ta polega na zastosowaniu średniej ruchomej z wykładniczo malejącymi wagami. Stosowana jest do usuwania szumu lub wizualizacji danych. Proste wygładzanie wykładnicze możemy matematycznie opisać za pomocą wzoru:

Okazuje się, że powyższy wzór możemy zapisać w postaci rekurencyjnej.

$$ s_t = (1-\alpha)s_{t-1} + \alpha x_t$$ 
$$s_1 = x_1 $$

# Identyfikacja modeli

**Twierdzenie** Dla procesu $AR(p)$ funkcja autokorelacji cząstkowej znika dla opóźnień większych niż rząd $p$, tzn. jeśli szereg $\{X_t\}$ jest procesem $AR(p)$, to
<center>
$ \alpha(h) \ne 0,$ jeżeli $h \le p $
<center>
$ \alpha(h) = 0,$ jeżeli $h > p $,

gdzie $\alpha(h)$ oznacza funkcję cząstkowej autokorelacji (PACF).

**Twierdzenie** Jeżeli szereg $\{X_t\}$ jest procesem $MA(q)$, to jego teoretyczna funkcja autokorelacji jest postaci
$$
\rho(h) = \left\{ \begin{array}{ll}
\sum_{j=0}^{q-|h|} \theta_j\theta_{j+|h|}/(1 + \theta_1^2 + \dots + \theta_q^2)& \textrm{jeżeli $|h| \le q,$ }\\
0 & \textrm{jeżeli $|h|>q,$}\\
\end{array} \right.
$$
gdzie przyjmujemy, że $\theta_0 = 1$.

# Reguły dopasowywania modeli do danych rzeczywistych

### Reguła identyfikacji białego szumu

Szereg możemy uznać za realizację białego szumu, jeżeli:
- co najmniej 95% autokorelacji próbokowych ($ACF(h), h = 1, 2, 3, \dots, h_{max}$) znajduje się w przedziale ufności: $\pm \frac{1.96}{\sqrt{n}}$
- nie ma autokorelacji "istotnie" wychodzących poza przedziały ufności $\pm \frac{1.96}{\sqrt{n}}$

```python
plot_acf(np.random.randn(100))
```

Na powyższym wykresie widzimy wartości funkcji ACF dla białego szumu. Czerwonym polem jest zaznaczony powyższy przedział ufności.

Powyższa reguła jest często także nazywana graficznym testem białoszumowości, ponieważ w bardzo łatwy sposób, graficznie, jesteśmy w stanie ocenić czy szereg lub jego pewna część zachowuje się jak biały szum.

### Reguła identyfikacji modelu $MA(q)$

Jeżeli autkorelacja próbkowa ACF(h) znajduje się pomiędzy przedziałami ufności $\pm \frac{1.96}{\sqrt{n}}$ dla $h>q$, to spodziewamy się, że dane są realizacją procesu średniej ruchomej rzędu $q$.

```python
ma = MAsimulation(100, [0.5, 0.3, 0.4])
plot_acf(ma)
```

Na powyższym wykresie widzimy przykład działania reguły dla symulacyjnie wygenerowanego szeregu $MA(3)$.

### Reguła identyfikacji modelu $AR(p)$

Jeżeli cząstokowa autokorelacja próbkowa PACF(h) znajduje się pomiędzy przedziałami ufności $\pm \frac{1.96}{\sqrt{n}}$ dla $h>p$, to spodziewamy się, że dane są realizacją procesu autoregresji rzędu $p$.

```python
ar = ARsimulation(100, [0.5, -0.4, -0.3])
plot_pacf(ar)
```

Na powyższym wykresie widzimy przykład działania reguły dla symulacyjnie wygenerowanego szeregu $AR(3)$.
